```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_bnp = pd.read_csv('BNP.csv', skiprows=4)
```

```python
#Vill du testa andra länder än dom jag använt mig av, så ändrar du och/eller lägger till dessa i listan nedan.
countries_of_interest = ['Sweden', 'Germany', 'Brazil', 'Thailand', 'Uganda'] 

#Här anger vi länderna samt åren som vi vill ha data för. I detta fall 1990 t.o.m 2020.
columns_of_interest_bnp = ['Country Name'] + [str(year) for year in range(1990, 2021)]
df_bnp_filtered = df_bnp[df_bnp['Country Name'].isin(countries_of_interest)][columns_of_interest_bnp]
```

```python
# Här kommer min "Data-cleaning" för BNP. Jag valde att ha denna i en och samma cell, för att underlätta rengöringsprocessen och köra denna på en och samma gång, istället för flera celler.

print("Saknade värden före rengöring i BNP-data:")
print(df_bnp_filtered.isnull().sum())

for year in columns_of_interest_bnp[1:]: 
    mean_value = df_bnp_filtered[year].mean()
    df_bnp_filtered[year].fillna(mean_value, inplace=True)

print("\nSaknade värden efter rengöring i BNP-data:")
print(df_bnp_filtered.isnull().sum())

print("\nDatastatistik efter rengöring för BNP:")
df_bnp_filtered.describe()
```

```python
plt.figure(figsize=(10, 8))  # Ange storleken på din figur

# Loopa genom varje land och plotta deras BNP data
for country in countries_of_interest:
    country_data = df_bnp_filtered[df_bnp_filtered['Country Name'] == country].iloc[0, 1:].astype(float)  # KHär konverterar jag till float.
    plt.plot(df_bnp_filtered.columns[1:], country_data, label=country)

plt.title('BNP per Capita 1990-2020 (valuta US$)')
plt.xlabel('År')
plt.ylabel('BNP per Capita (valuta US$)')
plt.xticks(rotation=45) 
plt.legend()
plt.show()
```

```python
countries_of_interest = ['Sweden', 'Germany', 'Brazil', 'Thailand', 'Uganda']  # Samma länder som tidigare

# Här väljer vi, som ovan, åren mellan 1990 t.o.m 2020
columns_of_interest_bnp_capita = ['Country Name'] + [str(year) for year in range(1990, 2021)]
df_bnp_per_capita_filtered = df_bnp_per_capita[df_bnp_per_capita['Country Name'].isin(countries_of_interest)][columns_of_interest_bnp_capita]
```

```python
df = pd.read_csv('forvantad_livslangd_GPD.csv', skiprows=4)
```

```python
df_filtered = df[(df['Country Name'].isin(countries_of_interest)) & (df['Indicator Name'] == 'Life expectancy at birth, total (years)')]

#Som ovan kan  du välja de länder som är av intresse.
columns_of_interest = ['Country Name'] + [str(year) for year in range(1990, 2021)]
df_filtered = df_filtered[columns_of_interest]
```

```python
#Vi kör även denna datacleaning i en och samma cell för den förväntade livslängden.

print("Saknade värden före rengöring:")
print(df_filtered.isnull().sum())

for country in countries_of_interest:
    for year in columns_of_interest[1:]:
        median_value = df_filtered[df_filtered['Country Name'] == country][year].median()
        df_filtered.loc[df_filtered['Country Name'] == country, year] = df_filtered[df_filtered['Country Name'] == country][year].fillna(median_value)

for year in columns_of_interest[1:]:
    df_filtered[year] = pd.to_numeric(df_filtered[year], errors='coerce')

print("\nDatastatistik efter rengöring:")
df_filtered.describe()

print("\nSaknade värden efter rengöring:")
print(df_filtered.isnull().sum())
```

```python
print(df_filtered)
```

```python
plt.figure(figsize=(10, 8))

for country in countries_of_interest:
    plt.plot(df_filtered.columns[1:], df_filtered[df_filtered['Country Name'] == country].iloc[0, 1:], label=country)

plt.title('Förväntad Livslängd 1990-2020')
plt.xlabel('År')
plt.ylabel('Förväntad Livslängd vid Födseln (år)')
plt.xticks(rotation=45)  
plt.legend()
plt.show()
```

```python
#FRÅGA 5

#Jag skapar ett diagram för se hur ekonomi och livslängd kan relaterade. Jag återanvänder även datan från de tidigare cellerna.

df_bnp_2020 = df_bnp_per_capita_filtered[['Country Name', '2020']].rename(columns={'2020': 'BNP per Capita 2020'})
df_life_2020 = df_filtered[['Country Name', '2020']].rename(columns={'2020': 'Expected Life Span 2020'})

df_merged_2020 = pd.merge(df_bnp_2020, df_life_2020, on='Country Name')

df_merged_2020['BNP per Capita 2020'] = pd.to_numeric(df_merged_2020['BNP per Capita 2020'], errors='coerce')
df_merged_2020['Expected Life Span 2020'] = pd.to_numeric(df_merged_2020['Expected Life Span 2020'], errors='coerce')

plt.figure(figsize=(10, 6))
plt.scatter(df_merged_2020['BNP per Capita 2020'], df_merged_2020['Expected Life Span 2020'], color='blue')

plt.title('Samband mellan BNP per Capita och Förväntad Livslängd 2020')
plt.xlabel('BNP per Capita 2020 (US$)')
plt.ylabel('Förväntad Livslängd 2020 (år)')

for i, txt in enumerate(df_merged_2020['Country Name']):
    plt.annotate(txt, (df_merged_2020['BNP per Capita 2020'].iloc[i], df_merged_2020['Expected Life Span 2020'].iloc[i]))

plt.show()

plt.show()
```

```python
#FRÅGA 5 FORTS.

mean_life_expectancy = df_filtered.iloc[:, 1:].mean().mean()
std_life_expectancy = df_filtered.iloc[:, 1:].std().std()

countries_above_std = df_filtered[df_filtered.iloc[:, 1:].mean(axis=1) > mean_life_expectancy + std_life_expectancy]['Country Name'].unique()

print("Länder med förväntad livslängd en standardavvikelse över medelvärdet är:")
for country in countries_above_std:
    print(country)

#Jag kollade på sambandet mellan pengar och livslängd genom att använda ett diagram. 
#Det verkar som att länder med mer pengar har folk som lever längre. Antagligen för att de har bättre sjukvård och mat. 
#Men bara för att två saker verkar relaterade betyder det inte att den ena orsakar den andra. Det finns ju andra faktorer som utbildning och var man bor som också spelar roll.
```

```python
#FRÅGA 6
#Resultat: Tomt,  detta förmodar jag pga länder med låg BNP inte har säkrskilt lång livslängd, enligt deländer jag använt mig av.

# Beräkna medelvärdet för BNP per capita och förväntad livslängd för år 2020
median_bnp_per_capita_2020 = df_bnp_per_capita_filtered['2020'].median()

# Antag att vi använder 'df_filtered' för data om förväntad livslängd, och inte df_life_expectancy_2020
median_life_expectancy_2020 = df_filtered['2020'].median()

# Identifiera länder med hög förväntad livslängd men låg BNP per capita för år 2020
# Antag att df_filtered redan är filtrerad för de länder och år vi är intresserade av
high_life_low_bnp_countries = df_filtered[
    (df_filtered['2020'] > median_life_expectancy_2020) &
    (df_bnp_per_capita_filtered[df_bnp_per_capita_filtered['Country Name'].isin(countries_of_interest)]['2020'] < median_bnp_per_capita_2020)
]['Country Name']

print("Länder med hög förväntad livslängd men låg BNP per capita 2020:")
for country in high_life_low_bnp_countries:
    print(country)
```

```python
#FRÅGA 7

# Beräkna medianvärden för BNP och förväntad livslängd 2020
median_bnp_2020 = df_bnp_filtered['2020'].median()
median_life_expectancy_2020 = df_filtered['2020'].median()

print(f"Median BNP 2020: {median_bnp_2020}")
print(f"Median förväntad livslängd 2020: {median_life_expectancy_2020}")

# Identifiera länder med "stor" ekonomi, dvs. BNP över medianen
large_economies = df_bnp_filtered[df_bnp_filtered['2020'] > median_bnp_2020]['Country Name']

print("\nStora ekonomier 2020:")
print(large_economies.tolist())

# Kontrollera om dessa länder har en "lång" förväntad livslängd, dvs. över medianen
large_economies_long_life = df_filtered[
    (df_filtered['Country Name'].isin(large_economies)) & 
    (df_filtered['2020'] > median_life_expectancy_2020)
]['Country Name']

print("\nStora ekonomier med lång förväntad livslängd 2020:")
print(large_economies_long_life.tolist())
```

```python
#FRÅGA 8

# Beräkna medianvärdet för BNP per capita 2020
median_bnp_per_capita_2020 = df_bnp_per_capita_filtered['2020'].median()

print(f"Median BNP per Capita 2020: {median_bnp_per_capita_2020}")

# Identifiera länder med en "stor" ekonomi baserat på BNP per capita, dvs. över medianen
large_economies_per_capita = df_bnp_per_capita_filtered[df_bnp_per_capita_filtered['2020'] > median_bnp_per_capita_2020]['Country Name']

print("\nStora ekonomier baserat på BNP per Capita 2020:")
print(large_economies_per_capita.tolist())

# Kontrollera om dessa länder har en "lång" förväntad livslängd, dvs. över medianen
large_economies_per_capita_long_life = df_filtered[
    (df_filtered['Country Name'].isin(large_economies_per_capita)) & 
    (df_filtered['2020'] > median_life_expectancy_2020)
]['Country Name']

print("\nStora ekonomier (per capita) med lång förväntad livslängd 2020:")
print(large_economies_per_capita_long_life.tolist())
```
